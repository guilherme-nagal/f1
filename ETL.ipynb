```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
def verify_duplicated(df):
    if df.duplicated().sum() > 0:
        print("Dataframe contains duplicated rows")
        df.drop_duplicates(inplace=True)
        return df
    print("Dataframe NOT contains duplicated rows")
    return df
```

```python
df_circuits = pd.read_csv("./data/circuits.csv")
df_constructor_results = pd.read_csv("./data/constructor_results.csv")
df_constructor_standings = pd.read_csv("./data/constructor_standings.csv")
df_constructors = pd.read_csv("./data/constructors.csv")
df_driver_standings = pd.read_csv("./data/driver_standings.csv")
df_drivers = pd.read_csv("./data/drivers.csv")
df_lap_times = pd.read_csv("./data/lap_times.csv")
df_pit_stops = pd.read_csv("./data/pit_stops.csv")
df_qualifying = pd.read_csv("./data/qualifying.csv")
df_races = pd.read_csv("./data/races.csv")
df_results = pd.read_csv("./data/results.csv")
df_seasons = pd.read_csv("./data/seasons.csv")
df_sprint_results = pd.read_csv("./data/sprint_results.csv")
df_status = pd.read_csv("./data/status.csv")

list_dfs = [
    (df_circuits, "circuits"),
    (df_constructor_results, "constructor_results"),
    (df_constructor_standings, "constructor_standings"),
    (df_constructors, "constructors"),
    (df_driver_standings, "driver_standings"),
    (df_drivers, "drivers"),
    (df_lap_times, "lap_times"),
    (df_pit_stops, "pit_stops"),
    (df_qualifying, "qualifying"),
    (df_races, "races"),
    (df_results, "results"),
    (df_seasons, "seasons"),
    (df_sprint_results, "sprint_results"),
    (df_status, "status")
]
```

```python
for df in list_dfs:
    print(df[1])
    print("Number rows: ", df[0].shape[0])
    df = verify_duplicated(df[0])
    df.replace("\\N", np.nan, inplace=True)
    print(df.isnull().sum())
    print()
```

```python
df_circuits.drop(["url"], axis=1, inplace=True)
df_constructor_results.drop(["status"], axis=1, inplace=True)
df_constructors.drop(["url"], axis=1, inplace=True)
df_drivers.drop(["url"], axis=1, inplace=True)
df_races.drop(["fp1_date", "fp1_time", "fp2_date", "fp2_time", "fp3_date", "fp3_time", "quali_date", "quali_time", "sprint_date", "sprint_time"], axis=1, inplace=True)
```

```python
conn = sqlite3.connect('f1.sqlite3')
cursor = conn.cursor()

df_circuits.to_sql('circuits', con=conn, if_exists='replace', index=False)
df_constructor_results.to_sql('constructor_results', con=conn, if_exists='replace', index=False)
df_constructor_standings.to_sql('constructor_standings', con=conn, if_exists='replace', index=False)
df_constructors.to_sql('constructors', con=conn, if_exists='replace', index=False)
df_driver_standings.to_sql('driver_standings', con=conn, if_exists='replace', index=False)
df_drivers.to_sql('drivers', con=conn, if_exists='replace', index=False)
df_lap_times.to_sql('lap_times', con=conn, if_exists='replace', index=False)
df_qualifying.to_sql('qualifying', con=conn, if_exists='replace', index=False)
df_races.to_sql('races', con=conn, if_exists='replace', index=False)
df_results.to_sql('results', con=conn, if_exists='replace', index=False)
df_seasons.to_sql('seasons', con=conn, if_exists='replace', index=False)
df_sprint_results.to_sql('sprint_results', con=conn, if_exists='replace', index=False)
df_status.to_sql('status', con=conn, if_exists='replace', index=False)

cursor.close()
conn.close()
```
